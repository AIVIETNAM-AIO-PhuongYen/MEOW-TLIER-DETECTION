```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pathlib import Path    
data_dir = Path.cwd().resolve().parent/ "data"
data_dir.mkdir(parents=True, exist_ok=True)

# Load the dataset
df_clean = pd.read_csv(rf"{data_dir}\clean.csv")
df_clean.head()
```

```python
def plot_distribution(df, column):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

plot_distribution(df_clean, 'food_weight_g')
```

```python
plot_distribution(df_clean, 'sleep_hours')
```

```python
df_extreme = pd.read_csv(rf"{data_dir}\extreme.csv")
df_extreme.head()
```

```python
plot_distribution(df_extreme, 'food_weight_g')
```

```python
plot_distribution(df_extreme, 'sleep_hours')
```

```python
df_noise = pd.read_csv(rf"{data_dir}\noisy.csv")
df_noise.head()
```

```python
plot_distribution(df_noise, 'food_weight_g')
```

```python
plot_distribution(df_noise, 'sleep_hours')
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scale_data(df, columns, method='standard'):
        scaler = StandardScaler() if method == 'standard' else MinMaxScaler()
        
        df_scaled = df.copy()
        df_scaled[columns] = scaler.fit_transform(df[columns])
        
        return df_scaled

df_clean_scaled = scale_data(df_clean, ['food_weight_g', 'sleep_hours'], method='standard')
df_clean_scaled.head()
```

```python
plot_distribution(df_clean_scaled, 'food_weight_g')
```
